```python
import open3d as o3d
pts_new =np.load("3d.npy")

def np_to_pcd(xyz,index):
    """
    convert numpy array to point cloud object in open3d
    """
    xyz = xyz.reshape(-1,3)
    pcd = o3d.PointCloud()
    pcd.points = o3d.Vector3dVector(xyz)
    pcd.paint_uniform_color(color_instance[index])
    return pcd

pts_new = np.load('3.npy')
n_pc = pts_new.shape[0]
pcds_gt = o3d.PointCloud()
for i in range(n_pc):
    xyz = pts_new[i,:]
    current_pcd = np_to_pcd(xyz,i)
    pcds_gt += current_pcd
```
